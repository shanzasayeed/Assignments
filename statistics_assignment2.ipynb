Q2. Difference between Mean, Median, and Mode:
Mean: The mean is the average value of a dataset calculated by summing all values and dividing by the number of values. It is sensitive to outliers.
Median: The median is the middle value of a dataset when it is arranged in ascending or descending order. It is not affected by extreme values (outliers).
Mode: The mode is the value that occurs most frequently in a dataset. It can be used for both qualitative and quantitative data.
These measures are used to measure the central tendency of a dataset by providing information about where the center or typical value lies.

Q5. Measures of Dispersion (Range, Variance, Standard Deviation):
Measures of dispersion such as range, variance, and standard deviation quantify the spread or variability of data points in a dataset. Range is the difference between the maximum and minimum values, variance measures the average squared deviation from the mean, and standard deviation is the square root of variance, providing a measure of average deviation from the mean.

### Q8. Skewness in Data:
Skewness in data refers to the asymmetry of the data distribution. A dataset can be:
- **Symmetric:** No skewness, the distribution is balanced around the mean.
- **Right Skewed (Positive Skewness):** The tail of the distribution extends to the right, and the mean > median.
- **Left Skewed (Negative Skewness):** The tail of the distribution extends to the left, and the mean < median.

### Q9. Position of Median in Right Skewed Data:
In right-skewed data, the tail extends to the right, and the mean is greater than the median. So, the position of the median with respect to the mean is:
\[ \text{Mean} > \text{Median} \]

### Q10. Covariance vs. Correlation:
- **Covariance:** Covariance measures the degree to which two variables change together. A positive covariance indicates a direct relationship, while a negative covariance indicates an inverse relationship. However, it does not provide the strength of the relationship.
- **Correlation:** Correlation is a standardized measure that quantifies the strength and direction of the linear relationship between two variables. It ranges from -1 to 1, where 1 indicates a perfect positive correlation, -1 indicates a perfect negative correlation, and 0 indicates no linear correlation.

These measures are used in statistical analysis to understand relationships between variables and assess their strength and direction.

### Q11. Formula for Sample Mean:
The formula for calculating the sample mean (average) is:
\[ \text{Mean} = \frac{\sum_{i=1}^{n} X_i}{n} \]
where \( X_i \) are the individual data points and \( n \) is the number of data points.

Example calculation:
For a dataset {10, 15, 20, 25, 30}, the mean is:
\[ \text{Mean} = \frac{10 + 15 + 20 + 25 + 30}{5} = \frac{100}{5} = 20 \]

### Q12. Relationship in Normal Distribution:
In a normal distribution:
- The mean, median, and mode are all equal.
- The distribution is symmetric around the mean.

### Q13. Difference between Covariance and Correlation:
- Covariance measures the strength and direction of the relationship between two variables but does not provide the scale of the relationship.
- Correlation is a standardized measure that not only quantifies the strength and direction but also provides the scale of the linear relationship between two variables.

### Q14. Outliers' Impact on Measures:
- Outliers can significantly affect the mean, pulling it towards extreme values.
- Outliers have less impact on the median, as it is not affected by extreme values.
- Outliers can increase the variability and influence the range, variance, and standard deviation.
